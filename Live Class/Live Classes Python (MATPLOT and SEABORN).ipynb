```python
Salesdata=[100,200,400,300,500]
Fruits=['Mango','Grapes','Lichi','Banana','Apple']

plt.xlabel('Fruits')
plt.ylabel('Salesdata')
plt.title('Bar Method')

#by default in bar method alignment is in the center and color is blue
plt.bar(Fruits,Salesdata, color='green',align="edge",)
plt.show()
```

**Distribution Plot**
* displot: Figure-level interface to distribution plot functions.
* histplot: Plot a histogram of binned counts with optional normalization or smoothing.
* jointplot: Draw a bivariate plot with univariate marginal disctributions.
* violinplot: Draw an enhanced boxplot usig kernel density estimation.
* rugplot: Plot marginal distribution by drawing ticks along the x znd y axis

### Categorical Distribution plot

the first is the familiar box plot(). This kind of plot shows the three quarantile values of the distribution along with extreme values. The "Whiskers" extend to point that lie within 1.5 IQRs of the lower and upper quarantile, and then observations that fall outside this range are displayed independently. This means that each value in the boxplot corresponds to an actual observation in the data.

```python
sns.boxplot(x='day', y='total_bill',data=tips)
```

```python
sns.boxplot(x='day', y='total_bill',hue='smoker',data=tips, palette='coolwarm')
```
